```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x):
    return np.sin(x)+0.5*x
```

```python
def create_plot(x,y,styles,labels,axlabels):
    plt.figure(figsize=(20,6))
    for i in range(len(x)):
        plt.plot(x[i],y[i],styles[i],label = labels[i])
        plt.xlabel(axlabels[0])
        plt.ylabel(axlabels[1])
    plt.legend(loc=0)
```

```python
x=np.linspace(-2*np.pi,2*np.pi,50)
```

```python
create_plot([x],[f(x)],['b'],['f(x)'],['x','f(x)'])
```

## Regression

$$min_{\alpha{1}\cdots\alpha_{D}} \frac{1}{I}\sum_{i=1}^{I}\left(y_{i}-\sum_{d=1}^{D}\alpha_{d}\cdot\beta_{d}(x_{i})\right)$$

```python
res = np.polyfit(x,f(x),deg=1,full=True)
# res
ry = np.polyval(res[0],x)
create_plot([x,x],[f(x),ry],['b','r.'],['f(x)','regress'],['x','lin_regr'])
```

```python
reg = np.polyfit(x,f(x),deg=5)
ry = np.polyval(reg,x)
create_plot([x,x],[f(x),ry],['b','r.'],['f(x)','d5_reg'],['x','lin_regr'])
```

```python
reg = np.polyfit(x,f(x),deg=7)
ry = np.polyval(reg,x)
create_plot([x,x],[f(x),ry],['b','r.'],['f(x)','d7_reg'],['x','lin_regr'])
```

```python
np.allclose(f(x),ry,rtol=1e-01)
```
